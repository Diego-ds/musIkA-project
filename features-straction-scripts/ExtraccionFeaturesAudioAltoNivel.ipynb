```python
df = pd.read_csv('/content/drive/MyDrive/pdg/Datos/lyrics/EnEsSongsDataset.csv');
```

```python
df = df.drop(columns=['Unnamed: 0', 'Unnamed: 0.1', 'Unnamed: 0.1.1'])
```

```python
df.head()
```

```python
#se añaden las columnas en las cuales se guardarán las features
#danceability, 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms', 'time_signature'
df2 = df.assign(danceability=0.0,energy=0.0, key=0.0, loudness=0.0, 
                mode=0.0, speechiness=0.0, acousticness=0.0, instrumentalness=0.0,
                liveness=0.0, valence=0.0, tempo=0.0, duration_ms=0.0, time_signature=0)
```

```python
df2.head()
```

```python
error_ids = []
```

```python
for ind in df2.index:
  row = df2.iloc[ind]
  id = row["id"]
  try:
    features = sp.audio_features(id)[0]
    df2.at[ind,'danceability'] = features['danceability']
    df2.at[ind,'energy'] = features['energy']
    df2.at[ind,'key'] = features['key']
    df2.at[ind,'loudness'] = features['loudness']
    df2.at[ind,'mode'] = features['mode']
    df2.at[ind,'speechiness'] = features['speechiness']
    df2.at[ind,'acousticness'] = features['acousticness']
    df2.at[ind,'instrumentalness'] = features['instrumentalness']
    df2.at[ind,'liveness'] = features['liveness']
    df2.at[ind,'valence'] = features['valence']
    df2.at[ind,'tempo'] = features['tempo']
    df2.at[ind,'duration_ms'] = features['duration_ms']
    df2.at[ind,'time_signature'] = features['time_signature']
  except:
    print(ind)
    print(type(ind))
    error_ids.append(ind)
    pass

#danceability, 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms', 'time_signature'
```
