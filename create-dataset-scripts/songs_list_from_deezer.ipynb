```python
import deezer
from pydeezer import Deezer
from pydeezer import Downloader
from pydeezer.constants import track_formats
import time
```

```python
client = deezer.Client()
arl = "0e940153107314cfeac9a1d7db9b4e93187145140fc9c23060f9ef9884bdfdc9dc5ef7b6b61ff116216e742eea056d5795897334e4fc8260017e800c4feef80d6a5b104848a619246cb0c6caac100003b2f5e117ac9782a5eaafd3750d1700e0"
deezeer = Deezer(arl=arl)
user_info = deezeer.user
```

```python
download_dir = "/content/sample_data"
```

```python
#Cargar lista de canciones
lista = [["rap god","eminem"],["dare","gorillaz"]]

for song,artist in lista:
    tracks = client.search(song) #poner la lista de canciones en lugar de rapgod
    print(tracks.total)
    
    for t in tracks:
        tittle = t.title.lower() #se busca el titulo de la cancion
        track_artist = t.artist.name.lower() #se busca al autor
        
        if (tittle == song) and  (track_artist == artist): #se revisa si el titulo de la cancion y el nombre del autor coinciden
            ##se procede a descargar la cancion
            track_id = t.id
            start = time.time();
            track = deezeer.get_track(track_id)
            track["download"](download_dir, quality=track_formats.FLAC)
            end = time.time()
            print("time used:",end-start)
            break
```

```python
pop = client.get_genre(132)
rock = client.get_genre(152)
```

```python
pop_songs = []
pop_authors = []
```

```python
pop_number = 0
```

```python
pop_artists = pop.get_artists()

for artist in pop_artists:
  albums = artist.get_albums()
  time.sleep(1)
  for album in albums:
    if album.genre_id==pop.id:
      pop_number = pop_number+album.nb_tracks   
      tracks = album.tracks
      for track in tracks:
        pop_songs.append(track.title_short)
        pop_authors.append(track.artist.name)
      time.sleep(1)
```
